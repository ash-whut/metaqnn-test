```python
from matplotlib import pyplot as plt
import csv
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("cifar10_logs_2ep/replay_database.csv")  
print(df.head())
```

```python
max_indices = df.groupby('net')['accuracy_best_val'].idxmax()
result_df = df.loc[max_indices][['net', 'accuracy_best_val']].sort_index()
print(result_df.shape)
print(result_df.head())
```

```python
xticks = list(range(1, result_df.shape[0] + 1))
plt.figure(figsize=(1, 1))
plt.show()
fig= plt.figure(figsize=(10,8))
plt.plot(xticks, result_df['accuracy_best_val'].tolist())
plt.ylabel('Validation Accuracy')
plt.xlabel('Model Iteration')
plt.title("Validation accuracy vs Model Iteration")
plt.grid()
```
